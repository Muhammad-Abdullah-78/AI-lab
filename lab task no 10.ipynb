```python

import pandas as pd
import numpy as np
```

```python


student = pd.read_csv(
    r"C:\Users\DELL\Desktop\AI lab tasks (007)\student_marks_prediction_dataset.csv"
)

print(student.head())
```

```python


print("We have {} rows.".format(student.shape[0]))
print("We have {} columns.".format(student.shape[1]))
```

```python


print(np.sum(pd.isnull(student)))
```

```python


student = student.fillna(student.mean(numeric_only=True))

print(student.isnull().sum())
```

```python


print(student.dtypes)
```

```python


cat_columns = student.select_dtypes(['object']).columns

student[cat_columns] = student[cat_columns].apply(
    lambda x: pd.factorize(x)[0]
)

print(student.dtypes)
```

```python


student = student.astype(np.int64)

print(student.dtypes)
```
